### Import Libraries

```python
import time
import matplotlib.pyplot as plt
import random
```

### Sequential Search Algorithm

```python
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1
```

### Binary Search Algorithm

```python
def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

### Time Measurement function

```python
def measure_time(search_func, data, target):
    start_time = time.time()
    search_func(data, target)
    end_time = time.time()
    return (end_time - start_time) * 1_000_000
```

### Data

```python
def generate_data(size):
    return list(range(1, size + 1))  # [1, 2, ..., size]

# size data
sizes = [10, 100, 1000, 10000, 50_000, 100_000, 500_000, 1_000_000]
sequential_times = []
binary_times = []
```

### Experimenting

```python

for size in sizes:
    data = generate_data(size)
    target = size  # Cari elemen terakhir untuk pengujian "worst-case"
    
    # hitung waktu Sequential Search di setiap size
    seq_time = measure_time(sequential_search, data, target)
    sequential_times.append(seq_time)
    
    # hitung waktu Binary Search di setiap size
    bin_time = measure_time(binary_search, data, target)
    binary_times.append(bin_time)
```

Visualization

```python
plt.figure(figsize=(10, 6))
plt.plot(sizes, sequential_times, label="Sequential Search", marker='o')
plt.plot(sizes, binary_times, label="Binary Search", marker='s')
plt.title("Comparing Search Algorithm Complexities (Structured Data)")
plt.xlabel("Data Size")
plt.ylabel("Execution Time (microseconds)")
plt.legend() #
plt.grid()
plt.show()
```

#### **Hasil Analisis**
1. **Sequential Search**:
   - Waktu eksekusi algoritma **Sequential Search** menunjukkan peningkatan yang signifikan seiring dengan bertambahnya ukuran data.
   - Kompleksitas algoritma ini bersifat **linear** (*O(n)*), yang tercermin dari grafik waktu eksekusi yang naik tajam seiring dengan pertambahan ukuran data.
   - Pada ukuran data 50.000, waktu eksekusi tercatat sekitar **2.078 mikrodetik**. Pada ukuran data 1.000.000, waktu eksekusi meningkat menjadi **35.980 mikrodetik**.

2. **Binary Search**:
   - Waktu eksekusi untuk **Binary Search** hampir tidak berubah meskipun ukuran data meningkat. Hal ini menunjukkan bahwa algoritma **Binary Search** memiliki kompleksitas **logaritmik** (*O(log n)*), yang membuatnya jauh lebih efisien dibandingkan dengan Sequential Search, terutama untuk dataset besar.
   - Namun, pada pengujian ini, nilai waktu eksekusi untuk Binary Search tetap **0 mikrodetik** hingga ukuran data 1.000.000, yang bisa disebabkan oleh waktu komputasi yang sangat kecil (mungkin dibulatkan ke 0) atau pengaruh optimisasi sistem.

#### **Kesimpulan**
- **Sequential Search** mengalami peningkatan waktu eksekusi yang tajam seiring dengan bertambahnya ukuran data. Ini menunjukkan bahwa untuk data besar, algoritma ini menjadi sangat tidak efisien.
- **Binary Search** lebih efisien dengan kompleksitas logaritmik, meskipun pada data yang sangat besar, pengukurannya menunjukkan nilai yang sangat kecil dalam mikrodetik.
- Untuk dataset yang besar, **Binary Search** jauh lebih cepat daripada **Sequential Search**, dan lebih disarankan digunakan ketika data sudah terurut.
